Employee Performance Bonus Eligibility

```python
# Employee performance data
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest score
highest_score = max(employees.values())

# Step 2: Find all employees having that score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 3: Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

Search Query Keyword Analysis


```python
# Input query
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation
import string
for ch in string.punctuation:
    query = query.replace(ch, "")

# Step 3: Split into words
words = query.split()

# Step 4: Count frequency
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Step 5: Keep only repeated keywords
repeated_keywords = {}

for word, count in frequency.items():
    if count > 1:
        repeated_keywords[word] = count

# Output
print(repeated_keywords)
```

Sensor Data Validation

```python
# Sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid readings
valid_readings = []

# Loop through list using index
for hour in range(len(sensor_readings)):
    reading = sensor_readings[hour]

    # Check if reading is even (valid)
    if reading % 2 == 0:
        valid_readings.append((hour, reading))

# Output
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Email Domain Usage Analysis

```python
# Email list
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Count domains
domain_count = {}

for email in emails:
    domain = email.split("@")[1]   # extract domain
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Step 2: Calculate percentage
total_users = len(emails)

print("Domain Usage Percentage:")
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

Sales Spike Detection

```python
# Daily sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average
average_sales = sum(sales) / len(sales)

# Step 2: Define spike threshold (30% above average)
threshold = average_sales * 1.30

# Step 3: Detect spikes
for day in range(len(sales)):
    if sales[day] > threshold:
        print(f"Day {day + 1}: {sales[day]}")
```
